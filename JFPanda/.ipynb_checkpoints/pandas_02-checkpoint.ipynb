```python
path6 = "./exercise_data/wind.data"
import datetime
# 将数据作存储并且设置前三列为合适的索引
data = pd.read_table(path6, sep='\s+',parse_dates=[[0,1,2]])
data
```

```python
import datetime

def fix_century(x):
    year = x.year - 100 if x.year > 1987 else x.year
    return datetime.date(year, x.month,x.day)
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)
data.head()
```

```python
# 将日期设置为索引
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
data = data.set_index('Yr_Mo_Dy')
data
```

```python
# 对于每个location 一共有多少数据缺失-- 这里以列为单位计算
data.isnull().sum()
```

```python
# 对应每一个location，一共有多少完整的数据值  真的很神奇，shape[0]返回的是行数值，减法后竟然是统计列表

data.shape[0] - data.isnull().sum()
```

```python
data.min() # 先求列的平均值，再此基础上计算平均
```

```python
# 创建一个名为loc_stats的数据框去计算并存储每个location的风速最小值，最大值，平均值和标准差
loc_stats = pd.DataFrame()

loc_stats['min'] = data.min() # min
loc_stats['max'] = data.max() # max 
loc_stats['mean'] = data.mean() # mean
loc_stats['std'] = data.std() # standard deviations

loc_stats
```
